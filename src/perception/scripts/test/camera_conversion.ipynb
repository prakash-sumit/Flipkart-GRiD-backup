```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x_values = np.array([54, 64.6, 81, 97.5, 115.7])
y_values = np.array([0.1060, 0.1195, 0.1530, 0.180, 0.218])

# Performing linear regression
coefficients = np.polyfit(x_values, y_values, 1)
slope, intercept = coefficients

# Creating the equation string
equation = f'y = {slope:.7f}x + {intercept:.7f}'

# Plotting the data and the regression line
plt.scatter(x_values, y_values, label='Data points')
plt.plot(x_values, slope * x_values + intercept, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()

# Printing the equation
print("Equation:", equation)
```
